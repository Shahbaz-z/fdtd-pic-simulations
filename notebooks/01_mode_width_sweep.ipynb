# Simulation 1 — Effective index vs waveguide width

A **strip waveguide** guides light by total internal reflection at the Si/SiO2 interfaces. The fundamental TE mode effective index $n_{\mathrm{eff}}$ increases with core width as confinement improves. Above a certain width, higher-order modes appear — the classic single-mode vs multimode design trade-off for 220 nm SOI at 1550 nm.

This notebook uses the **ModeSolver** (eigenmode, local, no FDTD credits).

```python
import matplotlib.pyplot as plt
import numpy as np

from fdtd_pic.config import PROFILE_WIDTHS_UM, WIDTH_SWEEP
from fdtd_pic.modes import solve_modes, sweep_widths
from fdtd_pic.plotting import apply_style, save_figure

apply_style()
```

```python
widths_um, n_eff_te00 = sweep_widths(WIDTH_SWEEP)

fig, ax = plt.subplots()
ax.plot(widths_um * 1000, n_eff_te00, 'o-')
ax.set_xlabel('Width (nm)')
ax.set_ylabel(r'$n_{\mathrm{eff}}$ (mode 0)')
ax.set_title('Effective index vs width — 220 nm SOI @ 1550 nm')
save_figure(fig, '../assets/mode_neff_vs_width.png')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 3.5))
for ax, w in zip(axes, PROFILE_WIDTHS_UM):
    result = solve_modes(w)
    i0 = int(np.argmax(result.n_eff))
    result.mode_solver.plot_field(field_name="Ey", val="abs", mode_index=i0, ax=ax)
    ax.set_title(f'w = {int(w*1000)} nm')
fig.suptitle(r'|E_y| mode profiles (TE0)')
fig.tight_layout()
save_figure(fig, '../assets/mode_profiles_3widths.png')
plt.show()
```

## Key takeaway

$n_{\mathrm{eff}}$ rises with width; at ~700 nm a second mode becomes relevant (see the rightmost graph, pretty cool!). For single-mode operation at 1550 nm on 220 nm SOI, stay near **400–500 nm** strip width.
